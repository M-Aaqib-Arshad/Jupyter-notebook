## Order and Color for Plots

```python
# Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style=None,rc=None)

# Load dataset
ship = sns.load_dataset('titanic')
ship
# Change figure
plt.figure(figsize=(6,3))

#draw a line plot
sns.barplot(x="sex",y="alone",hue="who",data=ship,order=["female","male"],color = 'grey')
plt.title("Plot of Titanic alone Male and Females")
sns.set_style("dark")
plt.show()
```

## Estimator

### Mean 

```python
# Import Libraries
import seaborn as sns
import numpy
import matplotlib.pyplot as plt
# sns.set_style(style=None,rc=None)

# Load dataset
ship = sns.load_dataset('titanic')
ship
# Change figure
plt.figure(figsize=(6,3))

#draw a line plot
sns.barplot(x="class",y="fare",hue="sex",data=ship,ci=None,
            estimator= mean)
plt.title("Plot of Titanic alone Male and Females")
# sns.set_style("dark")
plt.show()
```

### For Median

```python
# Import Libraries
import seaborn as sns
import numpy
import matplotlib.pyplot as plt
# sns.set_style(style=None,rc=None)

# Load dataset
ship = sns.load_dataset('titanic')
ship
# Change figure
plt.figure(figsize=(6,3))

#draw a line plot
sns.barplot(x="class",y="fare",hue="sex",data=ship,ci=None,
            estimator= median)
plt.title("Plot of Titanic alone Male and Females")
# sns.set_style("dark")
plt.show()
```

### Color Saturation 

```python
# Import Libraries
import seaborn as sns
import numpy
import matplotlib.pyplot as plt
# sns.set_style(style=None,rc=None)

# Load dataset
ship = sns.load_dataset('titanic')
ship
# Change figure
plt.figure(figsize=(6,3))

#draw a line plot
sns.barplot(x="class",y="fare",hue="sex",data=ship,ci=None,
                 estimator=lambda x: x.mean(),saturation=1)
plt.title("Plot of Titanic alone Male and Females")
# sns.set_style("dark")
plt.show()
```
